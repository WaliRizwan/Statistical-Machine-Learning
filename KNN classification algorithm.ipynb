```python
import numpy as np
import pandas as pd
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
```

# 1- Generate data from the two classes specified by the parameters given above. 

```python
# create two centers 
GD_Center_1 = [4,11]
Cov_Matrix_1 = [[0.3, 0], [0, 0.3]]
```

```python
GD_Center_2 = [10,3]
Cov_Matrix_2 = [[1, 0], [0, 1]]
```

```python
Parameter_1 = np.random.multivariate_normal(GD_Center_1,Cov_Matrix_1,100)
Parameter_2 = np.random.multivariate_normal(GD_Center_2,Cov_Matrix_2,100)
```

```python
Dataset = np.concatenate((Parameter_1, Parameter_2), axis = 1)
DataFrame = pd.DataFrame(Dataset)
DataFrame
```

# 2- Plot the data points using a different symbol for each of the two classes.

```python
a1 = DataFrame.iloc[:,0]
z1 = DataFrame.iloc[:,1]
a2 = DataFrame.iloc[:,2]
z2 = DataFrame.iloc[:,3]
```

```python
fig = plt.figure(figsize=(10,8))
plt.scatter(a1,z1,  edgecolor='k', s=50, c = 'green', label = 'Parameter_1')
plt.scatter(a2,z2, edgecolor = 'k', s=50, c = 'blue', label = 'Parameter_2')
plt.xlabel('Parameter_1')
plt.ylabel('Parameter_2')
plt.style.use('seaborn')
plt.legend()
plt.title('Data_set');
```
